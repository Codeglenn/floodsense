```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
sys.path.append("..")

from app.services.preprocessor import FloodFeatureEngineer

df_raw = pd.read_csv("../data/processed/training_data.csv", parse_dates=["time"])
print(f"Shape: {df_raw.shape}")
print(f"Date range: {df_raw['time'].min()} to {df_raw['time'].max()}")
print(f"Regions: {df_raw['region_name'].unique()}")
df_raw.head()
```

```python
engineer = FloodFeatureEngineer()
df = engineer.engineer_features(df_raw)
df = engineer.create_labels(df)
df = df.dropna(subset=engineer.feature_columns)

print(f"After engineering: {df.shape}")
print(f"\nClass distribution:")
print(df["risk_level"].value_counts())
print(f"\nClass percentages:")
print((df["risk_level"].value_counts(normalize=True) * 100).round(2))
```

```python
order  = ["LOW", "MEDIUM", "HIGH", "CRITICAL"]
colors = ["#22c55e", "#eab308", "#f97316", "#ef4444"]
counts = df["risk_level"].value_counts().reindex(order)

fig, ax = plt.subplots(figsize=(8, 5))
bars = ax.bar(order, counts, color=colors, edgecolor="white", linewidth=1.5)
ax.bar_label(bars, fmt="{:,.0f}", padding=3)
ax.set_title("Flood Risk Class Distribution", fontsize=14, fontweight="bold")
ax.set_ylabel("Number of Hourly Observations")
ax.set_xlabel("Risk Level")
plt.tight_layout()
plt.savefig("class_distribution.png", dpi=150)
plt.show()
```

```python
corr_features = [
    "rainfall_24h", "rainfall_72h", "antecedent_precipitation_index",
    "soil_moisture_combined", "peak_intensity_24h",
    "humidity", "pressure_change_6h", "risk_label"
]
corr = df[corr_features].corr()

fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="RdYlGn",
            center=0, ax=ax, linewidths=0.5)
ax.set_title("Feature Correlation Matrix", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.savefig("correlation_matrix.png", dpi=150)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
for ax, feature in zip(axes, ["rainfall_24h", "rainfall_72h"]):
    for level, color in zip(order, colors):
        subset = df[df["risk_level"] == level][feature]
        subset = subset[subset < subset.quantile(0.99)]
        ax.hist(subset, bins=50, alpha=0.6, label=level,
                color=color, density=True)
    ax.set_title(f"{feature} by Risk Level")
    ax.set_xlabel("mm")
    ax.set_ylabel("Density")
    ax.legend()
plt.tight_layout()
plt.savefig("rainfall_distributions.png", dpi=150)
plt.show()
```

```python
nairobi = df[df["region_name"] == "Nairobi"].copy()
monthly = nairobi.resample("ME", on="time").agg({
    "rainfall_24h": "mean",
    "risk_label": "mean"
})

fig, ax1 = plt.subplots(figsize=(14, 5))
ax2 = ax1.twinx()
ax1.bar(monthly.index, monthly["rainfall_24h"], alpha=0.5,
        color="#3b82f6", label="Avg 24h Rainfall (mm)")
ax2.plot(monthly.index, monthly["risk_label"], color="#ef4444",
         linewidth=2, marker="o", label="Avg Risk Score")
ax1.set_ylabel("Avg 24h Rainfall (mm)", color="#3b82f6")
ax2.set_ylabel("Avg Risk Score (0-3)", color="#ef4444")
ax1.set_title("Nairobi: Monthly Rainfall vs Flood Risk", fontweight="bold")
plt.tight_layout()
plt.savefig("nairobi_seasonal.png", dpi=150)
plt.show()
```
